```python
#import statements
from src.CSPclass import CSP
from src.utils import parse_neighbors
from src.algorithms import *
```

```python
#CSP interpretation
variables = ["A", "B", "C", "D", "E", "empty"] #putting this here otherwise the CSP considers the seats as the variables
domain = { 'A': ["Seat_1", "Seat_2", "Seat_3", "Seat_4", "Seat_5", "Seat_6"],
           'B': ["Seat_1", "Seat_2", "Seat_3", "Seat_4", "Seat_5", "Seat_6"],
           'C': ["Seat_1", "Seat_2", "Seat_3", "Seat_4", "Seat_5", "Seat_6"],
           'D': ["Seat_1", "Seat_2", "Seat_3", "Seat_4", "Seat_5", "Seat_6"],
           'E': ["Seat_1", "Seat_2", "Seat_3", "Seat_4", "Seat_5", "Seat_6"],
           'empty': ["Seat_1", "Seat_2", "Seat_3", "Seat_4", "Seat_5", "Seat_6"],}
#making seats like this so its easier to read
'''
seats = "A: B; A: C; A: D; A: E; A: empty; " \
        "B: C; B: D; B: E; B: empty; " \
        "C: D; C: E; C: empty; " \
        "D: E; D: empty; " \
        "E: empty;"
'''
#nevermind this makes seats A LOT easier to read
seats = 'A: B C D E empty; B: C D E empty; C: D E empty; D: E empty; E: empty'
neighbors = parse_neighbors(seats)
#big constraint portion to make sure that the neighboring seats to B can only be empty or D (since problems with A, C, and E)
constraint = lambda Xi, x, Yi, y: (
    x!=y and
    not(x == "B" and y not in ["D", "empty"]) and
    not( y == "B" and x not in ["D", "empty"])
)
#making the actual CSP thing
lab7CSP = CSP(variables, domains=domain, neighbors=neighbors, constraints=constraint)
```
