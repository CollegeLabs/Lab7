```python
#CSP interpretation
variables = ["A", "B", "C", "D", "E", "empty"] #putting this here otherwise the CSP considers the seats as the variables
domain = dict(Seat_A = ["A", "B", "C", "D", "E", "empty"],
              Seat_B = ["A", "B", "C", "D", "E", "empty"],
              Seat_C = ["A", "B", "C", "D", "E", "empty"],
              Seat_D = ["A", "B", "C", "D", "E", "empty"],
              Seat_E = ["A", "B", "C", "D", "E", "empty"],
              Seat_F = ["A", "B", "C", "D", "E", "empty"])
seats = "Seat_A: Seat_B; Seat_A: Seat_C; Seat_B: Seat_D; Seat_C: Seat_E; Seat_F: Seat_E; Seat_F: Seat_D"
neighbors = parse_neighbors(seats)
#big constraint portion to make sure that the neighboring seats to B can only be empty or D (since problems with A, C, and E)
constraint = lambda Xi, x, Yi, y: not (
    x == "B" and y not in ["D", "empty"]
) and not (
    y == "B" and x not in ["D", "empty"]
)
#making the actual CSP thing
lab7CSP = CSP(variables=variables, domains=domain, neighbors=neighbors, constraints=constraint)
```
